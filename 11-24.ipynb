```python
def findCapital(text):
    cap_lets = []
    low_lets = []
    for i in text:
        if i.isupper():
            cap_lets.append(i)
        elif i.islower():
            low_lets.append(i)
    return len(cap_lets)
```

```python
data['text'].apply(lambda x: findCapital(x))
```
